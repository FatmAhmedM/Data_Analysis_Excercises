### Step 6. Repeat the two previous steps for a few other stocks, always creating a new dataframe: Tesla, IBM and Microsoft. (Ticker symbols TSLA, IBM and MSFT.)

```python

reader = web.AVTimeSeriesReader(symbols='TSLA',start=time_range[0],end=time_range[-1],api_key='DJU1R3ZAP5IURQM5')
Tesla = reader.read()
Tesla['stock'] = 'TSLA'
Tesla
```

### Step 7. Combine the four separate dataFrames into one combined dataFrame df that holds the information for all four stocks

```python
frames = [df_apple,Tesla,IBM,micro]
df = pd.concat(frames)
df
```

### Step 8. Shift the stock column into the index (making it a multi-level index consisting of the ticker symbol and the date).

```python
df.index = pd.to_datetime(df.index)
df
```

```python
re =df.set_index('stock', append=True)
```

```python
re
```

### Step 7. Create a dataFrame called vol, with the volume values.

```python
vol = pd.DataFrame(re['volume'])
vol
```

### Step 8. Aggregate the data of volume to weekly.
Hint: Be careful to not sum data from the same week of 2015 and other years.

```python
# v = df['stock'].to_list()
# nw = df.resample('w').sum()
# nw = nw.assign(stock=v)
# nw  
vol.reset_index('stock',inplace=True)
vol.index.name = 'date'
vol.reset_index('date',inplace=True)
vol['date'] = pd.to_datetime(vol['date'], format='%Y-%m-%d')
vol.groupby(['stock',pd.Grouper(key='date',freq='W')]).sum()
```

### Step 9. Find all the volume traded in the year of 2015

```python
df.loc[df.index.year == 2015,['volume','stock']].groupby('stock').sum()
```
